```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
surveyDF = pd.read_csv("studentSurvey.csv")

surveyDF.head()
```

```python
cleanedSurveyDF = surveyDF.drop("What is your age?", axis=1)
cleanedSurveyDF['What is your major category?'] = cleanedSurveyDF['What is your major category?'].str.replace('Data Science', 'STEM')
cleanedSurveyDF['What is your major category?'] = cleanedSurveyDF['What is your major category?'].str.replace('Health Science$', 'Health Science (includes psychology)', regex=True)
cleanedSurveyDF['What is your major category?'] = cleanedSurveyDF['What is your major category?'].str.replace('Communication', 'Arts/Humanities')
cleanedSurveyDF['What is your major category?'] = cleanedSurveyDF['What is your major category?'].str.replace('Health Science (includes psychology)','Health Science')


cleanedSurveyDF.head()
```

```python
majorDF = cleanedSurveyDF.groupby("What is your major category?")

plt.pie(x = majorDF.size(),
        labels=majorDF.size().index,
        autopct='%1.1f%%')

plt.show()
```
